```python
import string
string.punctuation
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def cleanText(text):
    textPuncRemoved = [char for char in text if char not in string.punctuation]
    textPuncRemovedJoin = ''.join(textPuncRemoved)
    textPuncRemovedNoStop = [word for word in textPuncRemovedJoin.split() if word.lower() not in stopwords.words('english')]
    return textPuncRemovedNoStop
```

```python
reviews_df_clean = reviews_df['verified_reviews'].apply(cleanText)
```

```python
print(reviews_df['verified_reviews'][5])
```

```python
print(reviews_df_clean[5])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = cleanText)
reviewsCountVectorizer = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(reviewsCountVectorizer.toarray())
```

```python
reviewsCountVectorizer.shape
```

```python
reviews = pd.DataFrame(reviewsCountVectorizer.toarray())
```

```python
X = reviews
```

```python
y = reviews_df['feedback']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test = NB_classifier.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_predict_test)
sns.heatmap(conf_matrix, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
from sklearn.linear_model import LogisticRegression 

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot = True)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot = True)

print(classification_report(y_test, y_pred))
```
